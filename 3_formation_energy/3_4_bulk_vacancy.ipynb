```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../final_metal_oxide_properties.xlsx', sheet_name='oxide_correlation')
```

```python
df.drop(columns=['t_peak_1','t_peak_2','t_peak_3','Unnamed: 9','select_oxides','vacancy_form_energy(eV)'], inplace=True)
```

```python
df.head()
```

```python
df = df[np.isfinite(df['e_bulk_vacancy'])]
```

```python
df
```

```python
from matplotlib import rcParams
from adjustText import adjust_text
rcParams['axes.titlepad'] = 15
plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(20,15), dpi=500)
x = list(df['formation__energy_per_oxygen_ev'])
y = list(df['e_bulk_vacancy'])
label = list(df['unique_oxides'])
plt.scatter(x, y, s=100, color='red')
plt.title('Bulk vacancy formation energy vs standard gibbs formation energy per oxygen atom')
plt.xlabel('Standard gibbs free energy of oxide formation per O atom(eV)')
plt.ylabel('Bulk vacancy formation energy(eV)')
plt.xlim(-6.75,-4.25)
plt.ylim(1,6)
#plt.axhline(y=500, color='red', linestyle='--') 
#plt.axvline(x=-4.7, color='red', linestyle='--') 
#plt.axvline(x=-6.2, color='red', linestyle='--') 
for i, label in enumerate(label):
    plt.annotate(label, (x[i], y[i]), textcoords='offset points', xytext=(0,10), ha='center')
```
