```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rcParams
from adjustText import adjust_text
rcParams['axes.titlepad'] = 15
plt.rcParams.update({'font.size': 18})
x = list(df['formation__energy_per_oxygen_ev'])
y = list(df['t_melt'])
label = list(df['unique_oxides'])
plt.figure(figsize=(20,15), dpi=500)
plt.scatter(x, y, s=100, color='green')
#plt.title('Formation energy vs melting temperature')
plt.xlabel('Standard gibbs free energy of oxide formation per O atom(eV)')
plt.ylabel('Melting temperature($^\circ$C)')
plt.axhline(y=500, color='red', linestyle='--') 
plt.axvline(x=-4.7, color='red', linestyle='--') 
plt.axvline(x=-6.2, color='red', linestyle='--') 
plt.text(x=-6.45, y=-107, s='-6.2',bbox=dict(facecolor='red', alpha=0.2) )
plt.text(x=-4.67, y=-107, s='-4.7',bbox=dict(facecolor='red', alpha=0.2) )
plt.text(x=-9.1, y=550,s='Melting temperature > 500($^\circ$C) ')
for i, label in enumerate(label):
    plt.annotate(label, (x[i], y[i]), textcoords='offset points', xytext=(0,10), ha='center')
```
