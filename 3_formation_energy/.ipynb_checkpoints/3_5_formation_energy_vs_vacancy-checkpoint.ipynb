```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../final_metal_oxide_properties.xlsx', sheet_name='formation_energy_corr')
```

```python
df.head()
```

```python
df = df.replace('-', np.nan)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,4),dpi=300)
from matplotlib import rcParams
from adjustText import adjust_text
rcParams['axes.titlepad'] = 15
plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(20,15), dpi=500)
x = list(df['formation_energy_per_o_atom (eV)'])
y = list(df['bulk_vacancy_formation_energy (eV)'])
label = list(df['oxides'])
plt.scatter(x, y, s=100, color='red')
plt.title('Bulk vacancy formation energy vs Bulk formation energy per oxygen atom')
plt.xlabel('Bulk formation energy per O atom(eV)')
plt.ylabel('Bulk vacancy formation energy(eV)')
#plt.xlim(-6.75,-4.25)
#plt.ylim(1,6)
#plt.axhline(y=500, color='red', linestyle='--') 
#plt.axvline(x=-4.7, color='red', linestyle='--') 
#plt.axvline(x=-6.2, color='red', linestyle='--') 
for i, label in enumerate(label):
    plt.annotate(label, (x[i], y[i]), textcoords='offset points', xytext=(0,10), ha='center')
```
