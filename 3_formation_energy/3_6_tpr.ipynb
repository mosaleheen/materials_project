```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../final_metal_oxide_properties.xlsx', sheet_name='formation_energy_corr')
```

```python
df.head()
```

```python
df.drop(columns=['activation_energy(kj/mole)','Phase change'], inplace=True)
```

```python
df.head()
```

```python
df = df.replace('-', np.nan)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(8,4),dpi=300)
from matplotlib import rcParams
from adjustText import adjust_text
rcParams['axes.titlepad'] = 15
plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(20,15), dpi=500)
x = list(df['surface_vacancy_formation_energy (eV)'])
y = list(df['tpr'])
label = list(df['oxides'])
plt.scatter(x, y, s=100, color='red')
plt.title('Reduction temperature vs surface vacancy formation energy')
plt.xlabel('Surface vacancy formation energy(eV)')
plt.ylabel('Reduction temperature($^\circ$C)')
for i, label in enumerate(label):
    plt.annotate(label, (x[i], y[i]), textcoords='offset points', xytext=(0,10), ha='center', color='black')
```
