```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rcParams
from adjustText import adjust_text
rcParams['axes.titlepad'] = 15
plt.rcParams.update({'font.size': 18})
x = list(df['formation__energy_per_oxygen_ev'])
y_onset = list(df['t_onset'])
y_peak = list(df['t_max'])
label = list(df['unique_oxides'])
plt.figure(figsize=(20,15), dpi=500)
plt.scatter(x, y_onset, s=100)
#plt.title('Formation energy vs onset reduction temperature')
plt.xlabel('Standard gibbs free energy of oxide formation per O atom(eV)')
plt.ylabel('Onset reduction temperature($^\circ$C)')
plt.axhline(y=800, color='red', linestyle='--') 
plt.axvline(x=-4.7, color='red', linestyle='--') 
plt.axvline(x=-6.2, color='red', linestyle='--') 
plt.text(x=-6.4, y=23, s='-6.2',bbox=dict(facecolor='green', alpha=0.2) )
plt.text(x=-4.68, y=23, s='-4.7',bbox=dict(facecolor='green', alpha=0.2) )
plt.text(x=-6.5, y=810,s='Maximum onset reduction temperature ~800($^\circ$C) ')
for i, label in enumerate(label):
    plt.annotate(label, (x[i], y_onset[i]), textcoords='offset points', xytext=(0,10), ha='center')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['axes.titlepad'] = 15
plt.rcParams.update({'font.size': 18})
x = list(df['formation__energy_per_oxygen_ev'])
y_onset = list(df['t_onset'])
y_peak = list(df['t_max'])
label = list(df['unique_oxides'])
plt.figure(figsize=(20,15), dpi=500)
plt.scatter(x, y_peak, color='red', s =100)
#plt.title('Formation energy vs peak reduction temperature')
plt.xlabel('Standard gibbs free energy of oxide formation per O atom(eV)')
plt.ylabel('Peak reduction temperature($^\circ$C)')
plt.axhline(y=800, color='blue', linestyle='--') 
plt.axvline(x=-4.7, color='blue', linestyle='--') 
plt.axvline(x=-6.2, color='blue', linestyle='--') 
plt.text(x=-6.4, y=60, s='-6.2',bbox=dict(facecolor='red', alpha=0.2) )
plt.text(x=-4.68, y=60, s='-4.7',bbox=dict(facecolor='red', alpha=0.2) )
plt.text(x=-6.5, y=810,s='Maximum peak reduction temperature ~800($^\circ$C) ')
for i, label in enumerate(label):
     plt.annotate(label, (x[i], y_peak[i]), textcoords='offset points', xytext=(0.0,10), ha='center')
```
