```python
from pymatgen import MPRester
from pymatgen import Structure
```

```python
api_key='3ZvdRbGJX2dwZIaljnX'
mpr = MPRester(api_key)
```

```python
## getting the calculation data for sorted list of 28 oxides

oxides=['Bi2O3','CdO','CoO','Co3O4','Eu2O3','FeO','Fe2O3','Fe3O4','GeO2','In2O3','MnO','Mn2O3','Mn3O4',
       'MoO2','MoO3','PbO','Pb3O4','SbO2','Sb2O3','SnO','SnO2','Sn5O6','TeO2','V2O5','V3O5','V3O7',
       'WO2','WO3']

data=[]
for oxide in oxides:
    data.append(mpr.query(criteria={'pretty_formula':oxide,'e_above_hull':0}, 
                 properties=['material_id','pretty_formula','e_above_hull','input.incar','input.kpoints']))
```

```python
# apparently, it's a dictionary within a nested list
for index in range(len(data)):
    print(data[index][0])
```

```python
# checking how to access the kpoints data
for index in range(len(data)):
    print(data[index][0]['pretty_formula'])
    print(data[index][0]['input.kpoints'])
```

```python
# checking how to access the incar data
for i in range(len(data)):
    print('incar_'+data[i][0]['pretty_formula'])
    for j in sorted(data[i][0]['input.incar'].items()):
        print(j)
    print('\n')
```

```python
# writing the incar data into each incar_oxide file in a sorted and nice form
for i in range(len(data)):
    with open('incar_'+data[i][0]['pretty_formula'].lower(), 'w') as f:
        for j in sorted(data[i][0]['input.incar'].items()):
            line = '\t= '.join(str(x) for x in j)
            f.write(line + '\n')
```

```python
# writing the kpoints data into each kpoints_oxide file, it required a type conversion from kpoint type to string
for i in range(len(data)):
    with open('kpoints_'+data[i][0]['pretty_formula'].lower(), 'w') as f:
        line = str(data[i][0]['input.kpoints'])
        f.write(line)
```
