```python
from pymatgen import MPRester
from pymatgen import Structure
```

```python
api_key='3ZvdRbGJX2dwZIaljnX'
mpr = MPRester(api_key)
```

```python
## getting the cif data for sorted list of 13 metals

metals=['Bi','Cd','Co','Fe','Ge','In','Mn','Mo','Sb','Sn','Te','V','W']

data=[]
for metal in metals:
    data.append(mpr.query(criteria={'pretty_formula':metal,'e_above_hull':0}, 
                 properties=['material_id','pretty_formula','e_above_hull','cif','input.incar','input.kpoints']))
```

```python
# apparently, it's a dictionary within a nested list
for index in range(len(data)):
    print(data[index][0])
```

```python
# checking how to access the cif data
for index in range(len(data)):
    print(data[index][0]['pretty_formula'])
    print(data[index][0]['cif'])
```

```python
# writing the cif data into each cif_oxide file, it required a type conversion from cif type to string
for i in range(len(data)):
    with open('cif_'+data[i][0]['pretty_formula'].lower()+'.cif', 'w') as f:
        line = str(data[i][0]['cif'])
        f.write(line)
```

```python
#checking poscars, incars, and kpoints
for i in range(len(data)):
    poscar = Structure.from_str(data[i][0]["cif"], fmt="cif").to(fmt="poscar")
    incar = data[i][0]["input.incar"]
    kpoints = data[i][0]["input.kpoints"]
    print(poscar)
    print(incar)
    print(kpoints)
```

```python
# writing the incar data into each incar_metal file in a sorted and nice form
for i in range(len(data)):
    with open('incar_'+data[i][0]['pretty_formula'].lower(), 'w') as f:
        for j in sorted(data[i][0]['input.incar'].items()):
            line = '= '.join(str(x) for x in j)
            f.write(line + '\n')
```

```python
# writing the kpoints data into each kpoints_metal file, it required a type conversion from kpoint type to string
for i in range(len(data)):
    with open('kpoints_'+data[i][0]['pretty_formula'].lower(), 'w') as f:
        line = str(data[i][0]['input.kpoints'])
        f.write(line)
```

```python
# writing poscar data into each poscar_metal file
for i in range(len(data)):
    with open('poscar_'+data[i][0]['pretty_formula'].lower(), 'w') as f:
        poscar = Structure.from_str(data[i][0]["cif"], fmt="cif").to(fmt="poscar")
        line = str(poscar)
        f.write(line)
```
