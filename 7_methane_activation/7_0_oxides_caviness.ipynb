```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../final_metal_oxide_properties.xlsx', sheet_name='oxides_caviness')
```

```python
df
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(dpi=500)

label = list(df['oxides'])
y_pos = np.arange(len(label))
thermodynamics = list(df['delG_reaction_873K'])
ax.barh(y_pos, thermodynamics, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(label)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Gibbs free energy of methane dissociation (eV) at 600$^\circ$C')
ax.set_title('Thermodynamics of methane dissociation')

plt.show()
```
